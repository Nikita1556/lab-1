#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_v(text):  #Функция подсчитывает количество гласных букв в строке.
  v = 'aeiou'  # Определение строки с гласными буквами
  count = 0  # Инициалиризация счетчика гласных
  for c in text.lower():  # перевод каждой буквы в нижний регистр
    if c in v:  # проверка буквы на гласность
      count += 1  # Увеличение счетчика, если да
  return count  # Возврат количества гласных

# Тесты
def test_count_v():
  assert count_v("aeiou") == 5  # Проверка что функция возвращает число 5
  assert count_v("kgdgds") == 0  # Проверка что функция вернет число 0
  assert count_v("Hamster") == 2  # Проверка что функция вернет число 2
  print("Тесты завершены успешно")

test_count_v()  # функция для запуска тестирования
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def u(t): #Функция проверки-встречается ли каждый символ в строке лишь один раз.

  c_counts = {}  # массив для хранении информации о количестве символов с строке
  for c in t:  # Для каждого символа в строке
    if c in c_counts:  # Проверка есть ли символ в массиве
      return False  # Если символ уже был-возврат False
    else:
      c_counts[c] = 1  # Если символ встретился впервые, добавляем его в массив в количестве 1
  return True  # Если все символы встретились только по раз-возврат True

# Тесты
def t():
  assert u("abcdefg") == True  # Проверка- функция возвращает True 
  assert u("akkdlerr") == False  # Проверка- функция возвращает False
  assert u("H32l2f1") == False  # Проверка- функция возвращает False
  assert u("Hamster") == True  # Проверка- функция возвращает True
  
  print("Тесты завершены успешно")  
t()  # Функция для тестирования
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def c(n): #Функция счета количества единичных бит в двоичной записи числа.
 
  count = 0  # Инициализация счетчика единичных бит
  while n:  # Цикл идет , пока n не станет равным нулю
    count += n & 1  # Проверка последнего бита: если он единичный-увеличение счетчика
    n >>= 1  # Сдвиг числа вправо на 1 бит для проверки следующего бита
  return count  # Возвращение количества единичных бит

# Тесты
def t():
  assert c(0) == 0  # Проверка, что функция возвращает 0 
  assert c(1) == 1  # Проверка, что функция возвращает 1 
  assert c(6) == 2  # Проверка, что функция возвращает 2
  assert c(20) == 2 # Проверка, что функция возвращает 2 
  print("Тесты завершены успешно")

t()  # Функция для тестирования

```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):  # Функция подсчета количества операций умножения цифр, необходимых для получения однозначного числа.
  count = 0  # Инициализация счетчика операций
  while n > 9:  # Цикл продолжается, пока число не станет однозначным
    count += 1  # Увеличение счетчика операций
    n = p(n)  # Вычисление произведения цифр текущего числа
  return count  # Возвращение количества операций

def p(n):  # Функция вычисления произведения цифр в числе.
  b = 1  # Инициализация произведения цифр
  while n:
    b *= n % 10  # Умножение на последнюю цифру числа
    n //= 10  # Обрезание последней цифры числа
  return b  # Возвращение произведения цифр

# Тесты
def t_magic():
  assert magic(39) == 3  # Проверка, что функция возвращает 1 
  assert magic(42) == 1  # Проверка, что функция возвращает 1 
  assert magic(999) == 4  # Проверка, что функция возвращает 4 
  print("Тесты завершены успешно")

t_magic() # Функция для тестирования

```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
import numpy as np

def mse(p, t):  #Функция вычисляет среднеквадратическое отклонение между двумя векторами.
 
  return np.mean(np.square(p - t))  # Вычисляем среднеквадратического отклонения с помощью модуля NumPy

# Тесты
def t():
  p1 = np.array([4, 5, 6])
  t1 = np.array([4, 5, 6])
  assert mse(p1, t1) == 0.0  # Проверка, что среднеквадратического отклонение равно 0 для одинаковых векторов
  p2 = np.array([6, 5, 8])
  t2 = np.array([7, 6, 9])
  assert mse(p2, t2) == 1.0  # Проверка, что среднеквадратического отклонение равно 0 для векторов с разницей значений в еденицу
  p3 = np.array([1, 2, 3])
  t3 = np.array([5, 6, 7])
  assert mse(p3, t3) == 16.0  # Проверка, что среднеквадратического отклонение равно 0 для векторов с разницей в 4

  print("Тесты завершены успешно")

t() # Функция для тестирования

```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def f(n): # Функция раcкладывает целое число на простые множители и возвращает строку в виде "(n1p1)(n2p2)...(nkpk)".
  r = []  # Инициализиция пустого массива для хранения разложений
  for p in range(2, int(n**0.5) + 1):  # Проверка делимость на простые числа до квадратного корня из n
    count = 0  # Счетчик для степени простого числа
    while n % p == 0:  # Пока число делится на текущее простое число
      count += 1  # Увеличение счетчика
      n //= p  # Деление числа на простое число
    if count > 0:  # Если степень простого числа больше 0
      r.append(f"({p}{'**' + str(count) if count > 1 else ''})")  # Добавление в список строки с множителем

  if n > 1:  # Если число не было полностью разложено 
    r.append(f"({n})")  # Добавление последнего множителя

  return ''.join(r)  # Соединение всех множителей в одной строке

# Тесты
def t():
  assert f(12) == "(2**2)(3)"  # Проверка, что функция возвращает "(2**2)(3)"
  assert f(410) == "(2)(5)(41)"  # Проверка, что функция возвращает "(2)(5)(41)" 
print("Тесты завершены успешно")

t() # Функция для тестирования

```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python

def pyramid(n):  #Функция проверки-можно ли построить пирамиду из заданного числа кубиков.
 
  sum = 0  # Инициализация суммы квадратов
  z = 1  # Инициализирация числа слоев
  while sum < n:  # Пока сумма квадратов меньше заданного числа кубиков
    sum += z**2  # Добавление квадрата текущего слоя к сумме
    z += 1  # Переход к следующему слою
  if sum == n:  # Если сумма квадратов равна заданному числу
    return z - 1  # Возвращание количества слоев
  else:
    return "It is impossible"  # Если пирамиду построить нельзя, сообщаем об этом

# Тесты
def t():
  assert pyramid(14) == 3  # Проверка, что для 14 кубиков пирамида строится (1^2 + 2^2 + 3^2 = 14)
  assert pyramid(56) == "It is impossible"  # Проверка, что для 56 кубиков пирамиду построить нельзя
  assert pyramid(125) == "It is impossible"  # Проверка, что для 125 кубиков пирамиду построить нельзя
  print("Тесты завершены успешно")

t() # Функция для тестирования
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def a(n):  # Функция проверки - является ли число сбалансированным.
  s = str(n)  # Преобразование числа в строку
  l = len(s)  # Определение длины строки
  m = l // 2  # Определение позициии средней цифры в числе
  left_sum = sum(int(d) for d in s[:m])  # Сумма цифр до средних
  right_sum = sum(int(d) for d in s[m + (l % 2):])  # Сумма цифр после средних
  return left_sum == right_sum  # Проверка, равны ли суммы

# Тесты
def t():
  assert a(1234) == False  # Проверка, что 1234 - несбалансированное число
  assert a(569032) == False  # Проверка, что 569032 - несбалансированное число
  assert a(12344321) == True  # Проверка, что 12344321 - сбалансированное число
  print("Тесты завершены успешно")

t() # Функция для тестирования
```
